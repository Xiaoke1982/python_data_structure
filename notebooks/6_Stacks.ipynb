## 1. The Stack ADT

A **stack** is used to store data such that the last item inserted is the first item removed. It is used to implement a **last-in first-out** (**LIFO**) type protocol.

## 2. Implementing the Stack

### 2.1 Using a Python List

```python
class Stack:

    def __init__(self):
        self._items = list()

    
    def __len__(self):
        return len(self._items)

    
    def isEmpty(self):
        return len(self) == 0

    
    def pop(self):
        assert not self.isEmpty(), "cannot pop from an empty stack"
        return self._items.pop()  #Remove and return item at index (default last).


    def peek(self):
        assert not self.isEmpty(), "cannot peek from an empty stack"
        return self._items[-1]


    def push(self, item):
        self._items.append(item)
```

### 2.2 Using a Linked List

```python
class LinkedStack:

    def __init__(self):
        self._top = None
        self._size = 0

    def __len__(self):
        return self._size

    def isEmpty(self):
        return len(self)

    def push(self, value):
        newNode = _StackNode(value)
        newNode.next = self._top
        self._top = newNode
        self._size += 1

    def pop(self):
        assert not self.isEmpty(), "cannot pop from an empty stack"
        node = self._top
        self._top = self._top.next
        self._size -= 1
        node.next = None
        return node._value

    def peek(self):
        assert not self.isEmpty(), "cannot peek from an empty stack"
        return self._top._value
        
    

class _StackNode:

    def __init__(self, value):
        self._value = value
        self.next = None
    
```
